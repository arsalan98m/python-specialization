```python
# open function returns an object it's a file object and assigning it
# to the variable fileref

fileref = open("olympics.txt","r")
contents = fileref.read()
print(contents[:100])
fileref.close()
```

```python
# readlines method returns list object 
fileref = open("olympics.txt","r")
lines = fileref.readlines()
print(lines[:4])
fileref.close()
```

```python
fileref = open("olympics.txt","r")
lines = fileref.readlines()
for lin in lines[:4]:
    print(lin.strip())
fileref.close()
```

```python
fileref = open("olympics.txt","r")
for lin in fileref:
    print(lin.strip())
fileref.close()
```

```python
# file object do not support slicing
fileref = open("olympics.txt","r")
for lin in fileref[:4]:
    print(lin.strip())
fileref.close()
```

```python
# printing length of lines
# how many lines
fileref = open("olympics.txt","r")
lines = fileref.readlines()
print(len(lines))
fileref.close()
```

```python
# printing total characters 
fileref = open("olympics.txt","r")
contents = fileref.read()
print(len(contents))
fileref.close()
```
